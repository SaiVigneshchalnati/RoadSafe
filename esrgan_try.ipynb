```python
%%writefile app.py
import streamlit as st
import os
import shutil
import glob
import cv2
from PIL import Image
from base64 import b64encode

st.title("📈 Real-ESRGAN Image/Video Enhancer")

# Setup folders
for folder in ['frames', 'enhanced_frames', 'results', 'single_image', 'uploads']:
    if os.path.exists(folder):
        shutil.rmtree(folder)
    os.makedirs(folder, exist_ok=True)

uploaded_file = st.file_uploader("Upload an image (.jpg/.png) or video (.mp4/.avi)", type=["jpg", "jpeg", "png", "mp4", "avi", "mov", "mkv"])

if uploaded_file is not None:
    input_file_path = os.path.join("uploads", uploaded_file.name)

    with open(input_file_path, "wb") as f:
        f.write(uploaded_file.getbuffer())

    st.success(f"✅ Uploaded file: {uploaded_file.name}")

    is_video = uploaded_file.name.lower().endswith(('.mp4', '.mov', '.avi', '.mkv'))
    is_image = uploaded_file.name.lower().endswith(('.jpg', '.jpeg', '.png'))

    if is_video:
        st.info("📹 Detected input: Video")
        vidcap = cv2.VideoCapture(input_file_path)
        fps = vidcap.get(cv2.CAP_PROP_FPS)
        frame_count = 0

        while True:
            success, image = vidcap.read()
            if not success:
                break
            frame_name = f"frames/frame_{frame_count:05d}.png"
            cv2.imwrite(frame_name, image)
            frame_count += 1

        st.success(f"✅ Extracted {frame_count} frames at {fps:.2f} FPS.")
        input_for_esrgan = 'frames'

    elif is_image:
        st.info("🖼️ Detected input: Image")
        new_file_name = uploaded_file.name.replace(' ', '_').replace('(', '').replace(')', '')
        shutil.move(input_file_path, f'single_image/{new_file_name}')
        input_for_esrgan = f'single_image/{new_file_name}'

    else:
        st.error("❌ Unsupported file format!")
        st.stop()

    # Enhance using Real-ESRGAN
    st.info("🚀 Running Real-ESRGAN...")

    os.system(f'python inference_realesrgan.py -n RealESRGAN_x4plus -i "{input_for_esrgan}" --outscale 4 --face_enhance')

    # Display Results
    if is_image:
        enhanced_images = glob.glob('results/*')
        if len(enhanced_images) == 0:
            st.error("❌ Enhancement failed: No enhanced images found.")
            st.stop()

        enhanced_image_path = enhanced_images[0]
        st.image(Image.open(enhanced_image_path), caption="Enhanced Image", use_container_width=True)


        with open(enhanced_image_path, "rb") as file:
            btn = st.download_button(
                label="📥 Download Enhanced Image",
                data=file,
                file_name="enhanced_image.png",
                mime="image/png"
            )

    elif is_video:
        enhanced_images = sorted(glob.glob('results/*_out.png'))
        for img_path in enhanced_images:
            filename = os.path.basename(img_path).replace('_out', '')
            shutil.move(img_path, f"enhanced_frames/{filename}")

        os.system(f'ffmpeg -y -framerate {fps} -i enhanced_frames/frame_%05d.png -c:v libx264 -pix_fmt yuv420p enhanced_video.mp4')

        st.success("✅ Final enhanced video created: 'enhanced_video.mp4'")

        video_file = open('enhanced_video.mp4', 'rb')
        video_bytes = video_file.read()

        st.video(video_bytes)

        st.download_button(
            label="📥 Download Enhanced Video",
            data=video_bytes,
            file_name="enhanced_video.mp4",
            mime="video/mp4"
        )
```
