```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping
from pytorch_lightning.loggers import TensorBoardLogger
import shutil
import numpy as np
import matplotlib.pyplot as plt
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
from tqdm.auto import tqdm
import lpips
import glob
import random
from pathlib import Path
import zipfile
import gdown
import time
from datetime import datetime

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Create directories
os.makedirs("data", exist_ok=True)
os.makedirs("models", exist_ok=True)
os.makedirs("results", exist_ok=True)

# Step 1: Define Enhancement Goals
# We'll focus on:
# - Super-resolution (upscaling)
# - Noise reduction
# - Color enhancement
# - Sharpening

# Step 2: Data Collection & Preparation
class ImageEnhancementDataset(Dataset):
    def __init__(self, hr_images_dir, transform=None, degradation_transform=None,
                 img_size=256, scale_factor=4):
        self.hr_images_dir = hr_images_dir
        self.image_paths = glob.glob(os.path.join(hr_images_dir, "**/*.jpg"), recursive=True) + \
                         glob.glob(os.path.join(hr_images_dir, "**/*.png"), recursive=True)
        self.transform = transform
        self.degradation_transform = degradation_transform
        self.img_size = img_size
        self.scale_factor = scale_factor

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]

        # Read image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Create HR image
        if self.transform:
            hr_img = self.transform(image=img)['image']
        else:
            hr_img = img

        # Create LR image with degradations
        if self.degradation_transform:
            lr_img = self.degradation_transform(image=hr_img)['image']
        else:
            # Default degradation: downscale + mild noise
            h, w = hr_img.shape[:2]
            lr_h, lr_w = h // self.scale_factor, w // self.scale_factor
            lr_img = cv2.resize(hr_img, (lr_w, lr_h), interpolation=cv2.INTER_CUBIC)

            # Add noise
            noise = np.random.normal(0, 5, lr_img.shape).astype(np.uint8)
            lr_img = np.clip(lr_img + noise, 0, 255).astype(np.uint8)

            # Upscale back to original size for paired training
            lr_img = cv2.resize(lr_img, (w, h), interpolation=cv2.INTER_CUBIC)

        # Convert to tensors
        hr_tensor = transforms.ToTensor()(hr_img)
        lr_tensor = transforms.ToTensor()(lr_img)

        return {'lr': lr_tensor, 'hr': hr_tensor, 'path': img_path}

# Download DIV2K dataset (a commonly used dataset for image enhancement)
# def download_div2k():
#     print("Downloading DIV2K dataset...")

#     # URL for DIV2K validation set (smaller for quick testing)
#     url = "https://data.vision.ee.ethz.ch/cvl/DIV2K/DIV2K_valid_HR.zip"
#     output = "data/DIV2K_valid_HR.zip"

#     !wget -q {url} -O {output}

#     # Extract the dataset
#     with zipfile.ZipFile(output, 'r') as zip_ref:
#         zip_ref.extractall("data")

#     print("Dataset downloaded and extracted to data/DIV2K_valid_HR/")
#     return "data/DIV2K_valid_HR/"

import requests

def download_div2k():
    print("Downloading DIV2K dataset...")

    url = "https://data.vision.ee.ethz.ch/cvl/DIV2K/DIV2K_valid_HR.zip"
    output = "data/DIV2K_valid_HR.zip"

    if not os.path.exists(output):
        r = requests.get(url, stream=True)
        with open(output, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

    with zipfile.ZipFile(output, 'r') as zip_ref:
        zip_ref.extractall("data")

    print("Dataset downloaded and extracted to data/DIV2K_valid_HR/")
    return "data/DIV2K_valid_HR/"



# Define data transforms
def get_transforms(img_size=256):
    # HR image transform
    hr_transform = A.Compose([
        A.RandomCrop(height=img_size, width=img_size),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1, p=0.5),
    ])

    # LR image degradation transform (simulates various image problems)
    degradation_transform = A.Compose([
        # Downsample and upsample to lose information
        A.OneOf([
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_NEAREST, p=0.5),
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_LINEAR, p=0.5),
        ], p=1.0),

        # Add noise
        A.OneOf([
            A.GaussNoise(var_limit=(10.0, 50.0), p=0.5),
            A.ISONoise(color_shift=(0.01, 0.05), intensity=(0.1, 0.5), p=0.5),
            A.MultiplicativeNoise(multiplier=(0.9, 1.1), p=0.3),
        ], p=0.8),

        # Blur
        A.OneOf([
            A.GaussianBlur(blur_limit=(3, 7), p=0.5),
            A.MotionBlur(blur_limit=(3, 7), p=0.5),
        ], p=0.5),

        # JPEG compression artifacts
        A.ImageCompression(quality_lower=50, quality_upper=90, p=0.5),

        # Color distortions
        A.OneOf([
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
            A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
        ], p=0.5),
    ])

    return hr_transform, degradation_transform

# Data preparation class
class ImageEnhancementDataModule(pl.LightningDataModule):
    def __init__(self, data_dir, batch_size=8, img_size=256, num_workers=2):
        super().__init__()
        self.data_dir = data_dir
        self.batch_size = batch_size
        self.img_size = img_size
        self.num_workers = num_workers

    def setup(self, stage=None):
        hr_transform, degradation_transform = get_transforms(self.img_size)

        # Split into train and validation
        all_images = glob.glob(os.path.join(self.data_dir, "**/*.png"), recursive=True)
        if len(all_images) == 0:  # Try jpg if no png found
            all_images = glob.glob(os.path.join(self.data_dir, "**/*.jpg"), recursive=True)

        random.shuffle(all_images)
        split_idx = int(len(all_images) * 0.8)

        # Create train and validation directories
        os.makedirs("data/train", exist_ok=True)
        os.makedirs("data/val", exist_ok=True)

        # Create symlinks for train/val split
        for i, img_path in enumerate(all_images[:split_idx]):
            # os.symlink(os.path.abspath(img_path), f"data/train/{i}.jpg")

            shutil.copy2(img_path, f"data/train/{i}.jpg")



        for i, img_path in enumerate(all_images[split_idx:]):
            # os.symlink(os.path.abspath(img_path), f"data/val/{i}.jpg")
            shutil.copy2(img_path, f"data/val/{i}.jpg")

        self.train_dataset = ImageEnhancementDataset(
            "data/train",
            transform=hr_transform,
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

        self.val_dataset = ImageEnhancementDataset(
            "data/val",
            transform=A.Compose([A.RandomCrop(height=self.img_size, width=self.img_size)]),
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

    def train_dataloader(self):
        return DataLoader(
            self.train_dataset,
            batch_size=self.batch_size,
            shuffle=True,
            num_workers=self.num_workers,
            pin_memory=True
        )

    def val_dataloader(self):
        return DataLoader(
            self.val_dataset,
            batch_size=self.batch_size,
            shuffle=False,
            num_workers=self.num_workers,
            pin_memory=True
        )

# Step 3: Model Architecture Selection
# We'll implement a modified U-Net architecture with residual blocks and attention

# class ResidualBlock(nn.Module):
#     def __init__(self, channels):
#         super(ResidualBlock, self).__init__()
#         self.conv1 = nn.Conv2d(channels, channels, kernel_size=3, padding=1)
#         self.bn1 = nn.BatchNorm2d(channels)
#         self.relu = nn.ReLU(inplace=True)
#         self.conv2 = nn.Conv2d(channels, channels, kernel_size=3, padding=1)
#         self.bn2 = nn.BatchNorm2d(channels)

#     def forward(self, x):
#         residual = x
#         out = self.relu(self.bn1(self.conv1(x)))
#         out = self.bn2(self.conv2(out))
#         out += residual
#         out = self.relu(out)
#         return out

import torch.nn as nn

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels=None):
        if out_channels is None:
            out_channels = in_channels
        super(ResidualBlock, self).__init__()

        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)

        # Projection to match input to output shape if needed
        self.projection = nn.Sequential()
        if in_channels != out_channels:
            self.projection = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(out_channels)
            )

    # def forward(self, x):
    #     residual = self.projection(x)

    #     out = self.relu(self.bn1(self.conv1(x)))
    #     out = self.bn2(self.conv2(out))

    #     out += residual
    #     out = self.relu(out)

    #     return out
    def forward(self, x):
        residual = self.projection(x)  # Project input to match output dimensions if needed

        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))

        out += residual  # Add residual connection
        out = self.relu(out)

        return out



class ChannelAttention(nn.Module):
    def __init__(self, channels, reduction=16):
        super(ChannelAttention, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc = nn.Sequential(
            nn.Conv2d(channels, channels // reduction, 1, bias=False),
            nn.ReLU(inplace=True),
            nn.Conv2d(channels // reduction, channels, 1, bias=False)
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = self.fc(self.avg_pool(x))
        max_out = self.fc(self.max_pool(x))
        out = avg_out + max_out
        return self.sigmoid(out) * x

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size=kernel_size, padding=kernel_size//2)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        out = torch.cat([avg_out, max_out], dim=1)
        out = self.conv(out)
        return self.sigmoid(out) * x

class DownBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DownBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        # self.res_block = ResidualBlock(out_channels)
        self.res_block = ResidualBlock(in_channels=out_channels, out_channels=out_channels)

        self.channel_attention = ChannelAttention(out_channels)
        self.spatial_attention = SpatialAttention()

    def forward(self, x):
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        x = self.spatial_attention(x)
        return x

class UpBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UpBlock, self).__init__()
        self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        # self.conv = nn.Conv2d(in_channels + out_channels, out_channels, kernel_size=3, padding=1)
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(out_channels)
        self.channel_attention = ChannelAttention(out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        x = torch.cat([x2, x1], dim=1)

        x1 = self.relu(self.bn(self.conv(x1))) # Changed this line
        x = torch.cat([x2, x1], dim=1)


        x = self.res_block(x)
        x = self.channel_attention(x)
        return x

# class ImageEnhancerNet(nn.Module):
#     def __init__(self):
#         super(ImageEnhancerNet, self).__init__()
#         # Initial convolution to extract features
#         self.inc = nn.Sequential(
#             nn.Conv2d(3, 64, kernel_size=3, padding=1),
#             nn.ReLU(inplace=True)
#         )

#         # Encoder (downsampling) path
#         self.down1 = DownBlock(64, 128)
#         self.pool1 = nn.MaxPool2d(2)
#         self.down2 = DownBlock(128, 256)
#         self.pool2 = nn.MaxPool2d(2)
#         self.down3 = DownBlock(256, 512)
#         self.pool3 = nn.MaxPool2d(2)

#         # Bridge
#         # self.bridge = nn.Sequential(
#         #     ResidualBlock(512, 512),
#         #     ResidualBlock(512, 512),
#         #     ChannelAttention(512)
#         # )
#         self.bridge = nn.Sequential(
#             ResidualBlock(768, 256),  # in_channels=768, out_channels=256
#             ResidualBlock(256, 256),  # in_channels=256, out_channels=256
#             ChannelAttention(256)
#         )


#         # Decoder (upsampling) path
#         self.up1 = UpBlock(512, 256)
#         self.up2 = UpBlock(256, 128)
#         self.up3 = UpBlock(128, 64)

#         # Output convolution
#         self.outc = nn.Sequential(
#             nn.Conv2d(64, 64, kernel_size=3, padding=1),
#             nn.ReLU(inplace=True),
#             nn.Conv2d(64, 3, kernel_size=3, padding=1),
#             nn.Sigmoid()  # Output values between 0 and 1
#         )

#         # Global residual learning
#         self.global_res = nn.Conv2d(3, 3, kernel_size=1)

#     def forward(self, x):
#         # Save input for global residual connection
#         input_img = x

#         # Encoder path
#         x1 = self.inc(x)
#         x2 = self.down1(x1)
#         x2_pool = self.pool1(x2)

#         x3 = self.down2(x2_pool)
#         x3_pool = self.pool2(x3)

#         x4 = self.down3(x3_pool)
#         x4_pool = self.pool3(x4)

#         # Bridge
#         x5 = self.bridge(x4_pool)

#         # Decoder path
#         x = self.up1(x5, x4)
#         x = self.up2(x, x3)
#         x = self.up3(x, x2)

#         # Output
#         x = self.outc(x)

#         # Global residual connection (learn the residual)
#         residual = self.global_res(input_img)
#         output = torch.clamp(x + residual, 0, 1)

#         return output

class ImageEnhancerNet(nn.Module):
    def __init__(self):
        super(ImageEnhancerNet, self).__init__()
        # Initial convolution to extract features
        self.inc = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Encoder (downsampling) path
        self.down1 = DownBlock(64, 128)
        self.pool1 = nn.MaxPool2d(2)
        self.down2 = DownBlock(128, 256)
        self.pool2 = nn.MaxPool2d(2)
        self.down3 = DownBlock(256, 512)
        self.pool3 = nn.MaxPool2d(2)

        # Bridge (adjusting based on the previous block's output)
        self.bridge = nn.Sequential(
            ResidualBlock(512, 256),  # in_channels=512, out_channels=256 (change to 512 if needed)
            ResidualBlock(256, 256),  # in_channels=256, out_channels=256
            ChannelAttention(256)
        )

        # Decoder (upsampling) path
        self.up1 = UpBlock(512, 256)
        self.up2 = UpBlock(256, 128)
        self.up3 = UpBlock(128, 64)

        # Output convolution
        self.outc = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 3, kernel_size=3, padding=1),
            nn.Sigmoid()  # Output values between 0 and 1
        )

        # Global residual learning
        self.global_res = nn.Conv2d(3, 3, kernel_size=1)

    def forward(self, x):
        # Save input for global residual connection
        input_img = x

        # Encoder path
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x2_pool = self.pool1(x2)

        x3 = self.down2(x2_pool)
        x3_pool = self.pool2(x3)

        x4 = self.down3(x3_pool)
        x4_pool = self.pool3(x4)

        # Bridge
        x5 = self.bridge(x4_pool)

        # Decoder path
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)

        # Output
        x = self.outc(x)

        # Global residual connection (learn the residual)
        residual = self.global_res(input_img)
        output = torch.clamp(x + residual, 0, 1)

        return output


# Step 4: Training Infrastructure
class ImageEnhancementModule(pl.LightningModule):
    def __init__(self, lr=0.001):
        super().__init__()
        self.model = ImageEnhancerNet()
        self.lr = lr

        # Initialize LPIPS loss for perceptual quality
        self.lpips_loss = lpips.LPIPS(net='alex').to(self.device if hasattr(self, 'device') else 'cpu')

        # Save hyperparameters
        self.save_hyperparameters()

    def forward(self, x):
        return self.model(x)

    def training_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        # 1. Pixel-wise loss (L1)
        l1_loss = F.l1_loss(outputs, hr_imgs)

        # 2. Perceptual loss (LPIPS)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # 3. TV loss for smoothness
        tv_loss = self.total_variation_loss(outputs)

        # Combined loss
        loss = l1_loss + 0.1 * perceptual_loss + 0.0001 * tv_loss

        # Log losses
        self.log('train_l1_loss', l1_loss)
        self.log('train_perceptual_loss', perceptual_loss)
        self.log('train_tv_loss', tv_loss)
        self.log('train_total_loss', loss)

        return loss

    def validation_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        l1_loss = F.l1_loss(outputs, hr_imgs)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # Calculate metrics
        psnr_val = self.psnr(outputs, hr_imgs)
        ssim_val = self.ssim(outputs, hr_imgs)

        # Log metrics
        self.log('val_l1_loss', l1_loss)
        self.log('val_perceptual_loss', perceptual_loss)
        self.log('val_psnr', psnr_val)
        self.log('val_ssim', ssim_val)

        # Visualize results occasionally
        if batch_idx == 0:
            # Save sample images
            self.save_images(lr_imgs, outputs, hr_imgs)

        return {'val_loss': l1_loss, 'val_psnr': psnr_val, 'val_ssim': ssim_val}

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.lr)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=5, verbose=True
        )

        return {
            'optimizer': optimizer,
            'lr_scheduler': {
                'scheduler': scheduler,
                'monitor': 'val_l1_loss',
                'interval': 'epoch',
                'frequency': 1
            }
        }

    def total_variation_loss(self, x):
        # Total variation loss for smoother outputs
        batch_size = x.size()[0]
        h_x = x.size()[2]
        w_x = x.size()[3]
        count_h = self._tensor_size(x[:, :, 1:, :])
        count_w = self._tensor_size(x[:, :, :, 1:])
        h_tv = torch.pow((x[:, :, 1:, :] - x[:, :, :h_x-1, :]), 2).sum()
        w_tv = torch.pow((x[:, :, :, 1:] - x[:, :, :, :w_x-1]), 2).sum()
        return (h_tv + w_tv) / (batch_size * count_h * count_w)

    def _tensor_size(self, t):
        return t.size()[1] * t.size()[2] * t.size()[3]

    def psnr(self, img1, img2):
        # Peak Signal-to-Noise Ratio
        mse = F.mse_loss(img1, img2)
        if mse == 0:
            return 100
        return 20 * torch.log10(1.0 / torch.sqrt(mse))

    def ssim(self, img1, img2):
        # Simplified Structural Similarity Index
        C1 = 0.01**2
        C2 = 0.03**2

        img1 = img1.view(img1.size(0), -1)
        img2 = img2.view(img2.size(0), -1)

        mu1 = img1.mean(dim=1)
        mu2 = img2.mean(dim=1)

        sigma1 = (img1 - mu1.unsqueeze(1)).pow(2).mean(dim=1)
        sigma2 = (img2 - mu2.unsqueeze(1)).pow(2).mean(dim=1)
        sigma12 = ((img1 - mu1.unsqueeze(1)) * (img2 - mu2.unsqueeze(1))).mean(dim=1)

        mu1 = mu1.unsqueeze(1)
        mu2 = mu2.unsqueeze(1)

        ssim_value = ((2 * mu1 * mu2 + C1) * (2 * sigma12 + C2)) / \
                    ((mu1.pow(2) + mu2.pow(2) + C1) * (sigma1 + sigma2 + C2))

        return ssim_value.mean()

    def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
        # Create a grid of images
        num_samples = min(lr_imgs.size(0), max_samples)

        # Convert tensors to numpy arrays
        lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
        sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
        hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

        # Create figure
        plt.figure(figsize=(15, 5 * num_samples))

        for i in range(num_samples):
            # Original low-res
            plt.subplot(num_samples, 3, i*3 + 1)
            plt.imshow(lr_imgs_np[i])
            plt.title('Low Resolution')
            plt.axis('off')

            # Enhanced
            plt.subplot(num_samples, 3, i*3 + 2)
            plt.imshow(sr_imgs_np[i])
            plt.title('Enhanced')
            plt.axis('off')

            # Ground truth
            plt.subplot(num_samples, 3, i*3 + 3)
            plt.imshow(hr_imgs_np[i])
            plt.title('High Resolution')
            plt.axis('off')

        # Save figure
        os.makedirs('results', exist_ok=True)
        plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
        plt.close()

# Step 5: Evaluation Metrics
# (Already implemented in the training module)

# Main execution
def main():
    print("Starting image enhancement AI training...")

    # Download and prepare dataset
    data_dir = download_div2k()

    # Create data module
    data_module = ImageEnhancementDataModule(
        data_dir=data_dir,
        batch_size=4,  # Adjust based on your GPU memory
        img_size=256
    )

    # Initialize model
    model = ImageEnhancementModule(lr=0.0002)

    # Setup callbacks
    checkpoint_callback = ModelCheckpoint(
        dirpath='models',
        filename='image_enhancer-{epoch:02d}-{val_psnr:.2f}',
        save_top_k=3,
        monitor='val_psnr',
        mode='max'
    )

    early_stop_callback = EarlyStopping(
        monitor='val_psnr',
        patience=10,
        mode='max'
    )

    logger = TensorBoardLogger('logs', name='image_enhancer')

    # Initialize trainer
    trainer = pl.Trainer(
        accelerator='gpu' if torch.cuda.is_available() else 'cpu',
        devices=1,
        max_epochs=30,
        callbacks=[checkpoint_callback, early_stop_callback],
        logger=logger,
        log_every_n_steps=10,
        precision='16-mixed' if torch.cuda.is_available() else '32'  # Use mixed precision for faster training
    )

    # Train model
    print("Starting training...")
    trainer.fit(model, data_module)

    # Save final model
    torch.save(model.state_dict(), 'models/image_enhancer_final.pth')

    print(f"Training completed. Best model saved: {checkpoint_callback.best_model_path}")
    print(f"Best PSNR: {checkpoint_callback.best_model_score:.4f}")

    # Test the model on a few images
    test_model(model, data_module)

def test_model(model, data_module):
    """Test the trained model on a few sample images."""
    print("Testing the model on sample images...")

    # Load a batch from validation set
    val_dataloader = data_module.val_dataloader()
    test_batch = next(iter(val_dataloader))

    # Move to GPU if available
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    model.eval()

    lr_imgs = test_batch['lr'].to(device)
    hr_imgs = test_batch['hr'].to(device)

    # Generate enhanced images
    with torch.no_grad():
        enhanced_imgs = model(lr_imgs)

    # Calculate metrics
    psnr_values = []
    ssim_values = []

    for i in range(lr_imgs.size(0)):
        psnr_val = model.psnr(enhanced_imgs[i:i+1], hr_imgs[i:i+1]).item()
        ssim_val = model.ssim(enhanced_imgs[i:i+1], hr_imgs[i:i+1]).item()

        psnr_values.append(psnr_val)
        ssim_values.append(ssim_val)

    avg_psnr = sum(psnr_values) / len(psnr_values)
    avg_ssim = sum(ssim_values) / len(ssim_values)

    print(f"Test Results - Average PSNR: {avg_psnr:.4f}, Average SSIM: {avg_ssim:.4f}")

    # Save sample enhanced images
    os.makedirs('results/test', exist_ok=True)

    for i in range(min(4, lr_imgs.size(0))):
        # Convert tensors to numpy arrays
        lr_img = lr_imgs[i].cpu().detach().permute(1, 2, 0).numpy()
        enhanced_img = enhanced_imgs[i].cpu().detach().permute(1, 2, 0).numpy()
        hr_img = hr_imgs[i].cpu().detach().permute(1, 2, 0).numpy()

        # Create comparison figure
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.imshow(lr_img)
        plt.title(f'Low Resolution')
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.imshow(enhanced_img)
        plt.title(f'Enhanced (PSNR: {psnr_values[i]:.2f}, SSIM: {ssim_values[i]:.2f})')
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.imshow(hr_img)
        plt.title('High Resolution (Ground Truth)')
        plt.axis('off')

        plt.tight_layout()
        plt.savefig(f'results/test/sample_{i}.png')
        plt.close()

if __name__ == "__main__":
  main()

# if __name__ == "__main__":
#     data_path = download_div2k()

#     datamodule = ImageEnhancementDataModule(data_dir=data_path)
#     model = ImageEnhancementModule()

#     checkpoint_callback = ModelCheckpoint(monitor="val_loss", save_top_k=1, mode="min")
#     early_stop = EarlyStopping(monitor="val_loss", patience=5, verbose=True)
#     logger = TensorBoardLogger("lightning_logs", name="enhancer")

#     trainer = pl.Trainer(
#         max_epochs=5,
#         accelerator="auto",
#         callbacks=[checkpoint_callback, early_stop],
#         logger=logger,
#     )

#     trainer.fit(model, datamodule)
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping
from pytorch_lightning.loggers import TensorBoardLogger
import shutil
import numpy as np
import matplotlib.pyplot as plt
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
from tqdm.auto import tqdm
import lpips
import glob
import random
from pathlib import Path
import zipfile
import gdown
import time
from datetime import datetime
import requests

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Create directories
os.makedirs("data", exist_ok=True)
os.makedirs("models", exist_ok=True)
os.makedirs("results", exist_ok=True)

# Step 1: Define Enhancement Goals
# We'll focus on:
# - Super-resolution (upscaling)
# - Noise reduction
# - Color enhancement
# - Sharpening

# Step 2: Data Collection & Preparation
class ImageEnhancementDataset(Dataset):
    def __init__(self, hr_images_dir, transform=None, degradation_transform=None,
                 img_size=256, scale_factor=4):
        self.hr_images_dir = hr_images_dir
        self.image_paths = glob.glob(os.path.join(hr_images_dir, "**/*.jpg"), recursive=True) + \
                         glob.glob(os.path.join(hr_images_dir, "**/*.png"), recursive=True)
        self.transform = transform
        self.degradation_transform = degradation_transform
        self.img_size = img_size
        self.scale_factor = scale_factor

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]

        # Read image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Create HR image
        if self.transform:
            hr_img = self.transform(image=img)['image']
        else:
            hr_img = img

        # Create LR image with degradations
        if self.degradation_transform:
            lr_img = self.degradation_transform(image=hr_img)['image']
        else:
            # Default degradation: downscale + mild noise
            h, w = hr_img.shape[:2]
            lr_h, lr_w = h // self.scale_factor, w // self.scale_factor
            lr_img = cv2.resize(hr_img, (lr_w, lr_h), interpolation=cv2.INTER_CUBIC)

            # Add noise
            noise = np.random.normal(0, 5, lr_img.shape).astype(np.uint8)
            lr_img = np.clip(lr_img + noise, 0, 255).astype(np.uint8)

            # Upscale back to original size for paired training
            lr_img = cv2.resize(lr_img, (w, h), interpolation=cv2.INTER_CUBIC)

        # Convert to tensors
        hr_tensor = transforms.ToTensor()(hr_img)
        lr_tensor = transforms.ToTensor()(lr_img)

        return {'lr': lr_tensor, 'hr': hr_tensor, 'path': img_path}

def download_div2k():
    print("Downloading DIV2K dataset...")

    url = "https://data.vision.ee.ethz.ch/cvl/DIV2K/DIV2K_valid_HR.zip"
    output = "data/DIV2K_valid_HR.zip"

    if not os.path.exists(output):
        r = requests.get(url, stream=True)
        with open(output, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

    # Create extraction directory if it doesn't exist
    os.makedirs("data/DIV2K_valid_HR", exist_ok=True)

    with zipfile.ZipFile(output, 'r') as zip_ref:
        zip_ref.extractall("data")

    print("Dataset downloaded and extracted to data/DIV2K_valid_HR/")
    return "data/DIV2K_valid_HR/"

# Define data transforms
def get_transforms(img_size=256):
    # HR image transform
    hr_transform = A.Compose([
        A.RandomCrop(height=img_size, width=img_size),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1, p=0.5),
    ])

    # LR image degradation transform (simulates various image problems)
    degradation_transform = A.Compose([
        # Downsample and upsample to lose information
        A.OneOf([
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_NEAREST, p=0.5),
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_LINEAR, p=0.5),
        ], p=1.0),

        # Add noise
        A.OneOf([
            A.GaussNoise(var_limit=(10.0, 50.0), p=0.5),
            A.ISONoise(color_shift=(0.01, 0.05), intensity=(0.1, 0.5), p=0.5),
            A.MultiplicativeNoise(multiplier=(0.9, 1.1), p=0.3),
        ], p=0.8),

        # Blur
        A.OneOf([
            A.GaussianBlur(blur_limit=(3, 7), p=0.5),
            A.MotionBlur(blur_limit=(3, 7), p=0.5),
        ], p=0.5),

        # JPEG compression artifacts
        A.ImageCompression(quality_lower=50, quality_upper=90, p=0.5),

        # Color distortions
        A.OneOf([
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
            A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
        ], p=0.5),
    ])

    return hr_transform, degradation_transform

# Data preparation class
class ImageEnhancementDataModule(pl.LightningDataModule):
    def __init__(self, data_dir, batch_size=8, img_size=256, num_workers=2):
        super().__init__()
        self.data_dir = data_dir
        self.batch_size = batch_size
        self.img_size = img_size
        self.num_workers = num_workers

    def setup(self, stage=None):
        hr_transform, degradation_transform = get_transforms(self.img_size)

        # Split into train and validation
        all_images = glob.glob(os.path.join(self.data_dir, "**/*.png"), recursive=True)
        if len(all_images) == 0:  # Try jpg if no png found
            all_images = glob.glob(os.path.join(self.data_dir, "**/*.jpg"), recursive=True)

        if len(all_images) == 0:
            raise ValueError(f"No images found in {self.data_dir}. Please check the path.")

        random.shuffle(all_images)
        split_idx = int(len(all_images) * 0.8)

        # Create train and validation directories
        os.makedirs("data/train", exist_ok=True)
        os.makedirs("data/val", exist_ok=True)

        # Create copies for train/val split
        for i, img_path in enumerate(all_images[:split_idx]):
            shutil.copy2(img_path, f"data/train/{i}.jpg")

        for i, img_path in enumerate(all_images[split_idx:]):
            shutil.copy2(img_path, f"data/val/{i}.jpg")

        self.train_dataset = ImageEnhancementDataset(
            "data/train",
            transform=hr_transform,
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

        self.val_dataset = ImageEnhancementDataset(
            "data/val",
            transform=A.Compose([A.RandomCrop(height=self.img_size, width=self.img_size)]),
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

    def train_dataloader(self):
        return DataLoader(
            self.train_dataset,
            batch_size=self.batch_size,
            shuffle=True,
            num_workers=self.num_workers,
            pin_memory=True
        )

    def val_dataloader(self):
        return DataLoader(
            self.val_dataset,
            batch_size=self.batch_size,
            shuffle=False,
            num_workers=self.num_workers,
            pin_memory=True
        )

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels=None):
        super(ResidualBlock, self).__init__()
        if out_channels is None:
            out_channels = in_channels

        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)

        # Projection to match input to output shape if needed
        self.projection = nn.Sequential()
        if in_channels != out_channels:
            self.projection = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(out_channels)
            )

    def forward(self, x):
        residual = self.projection(x)  # Project input to match output dimensions if needed

        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))

        out += residual  # Add residual connection
        out = self.relu(out)

        return out

class ChannelAttention(nn.Module):
    def __init__(self, channels, reduction=16):
        super(ChannelAttention, self).__init__()
        # Ensure reduction doesn't make channels too small
        reduction = min(reduction, max(1, channels // 4))

        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc = nn.Sequential(
            nn.Conv2d(channels, max(1, channels // reduction), 1, bias=False),
            nn.ReLU(inplace=True),
            nn.Conv2d(max(1, channels // reduction), channels, 1, bias=False)
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = self.fc(self.avg_pool(x))
        max_out = self.fc(self.max_pool(x))
        out = avg_out + max_out
        return self.sigmoid(out) * x

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size=kernel_size, padding=kernel_size//2)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        out = torch.cat([avg_out, max_out], dim=1)
        out = self.conv(out)
        return self.sigmoid(out) * x

class DownBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DownBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(in_channels=out_channels, out_channels=out_channels)
        self.channel_attention = ChannelAttention(out_channels)
        self.spatial_attention = SpatialAttention()

    def forward(self, x):
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        x = self.spatial_attention(x)
        return x

class UpBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UpBlock, self).__init__()
        self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        # Corrected: input channels should be in_channels + skip connection's channels
        self.conv = nn.Conv2d(in_channels + out_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(out_channels)
        self.channel_attention = ChannelAttention(out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # Now ensure x1 has the same spatial dimensions as x2
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                         diffY // 2, diffY - diffY // 2])

        # Concatenate along channel dimension
        x = torch.cat([x2, x1], dim=1)
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        return x

class ImageEnhancerNet(nn.Module):
    def __init__(self):
        super(ImageEnhancerNet, self).__init__()
        # Initial convolution to extract features
        self.inc = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Encoder (downsampling) path
        self.down1 = DownBlock(64, 128)
        self.pool1 = nn.MaxPool2d(2)
        self.down2 = DownBlock(128, 256)
        self.pool2 = nn.MaxPool2d(2)
        self.down3 = DownBlock(256, 512)
        self.pool3 = nn.MaxPool2d(2)

        # Bridge
        self.bridge = nn.Sequential(
            ResidualBlock(512, 512),  # Keep the channels consistent
            ResidualBlock(512, 512),
            ChannelAttention(512)
        )

        # Decoder (upsampling) path - fixed channel dimensions
        self.up1 = UpBlock(512, 512)  # Bridge output + skip connection from down3
        self.up2 = UpBlock(512, 256)  # up1 output + skip connection from down2
        self.up3 = UpBlock(256, 128)  # up2 output + skip connection from down1

        # Final layers to produce the output
        self.outc = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 3, kernel_size=3, padding=1),
            nn.Sigmoid()  # Output values between 0 and 1
        )

        # Global residual learning
        self.global_res = nn.Conv2d(3, 3, kernel_size=1)

    def forward(self, x):
        # Save input for global residual connection
        input_img = x

        # Encoder path
        x1 = self.inc(x)  # [B, 64, H, W]

        x2 = self.down1(x1)  # [B, 128, H, W]
        x2_pool = self.pool1(x2)  # [B, 128, H/2, W/2]

        x3 = self.down2(x2_pool)  # [B, 256, H/2, W/2]
        x3_pool = self.pool2(x3)  # [B, 256, H/4, W/4]

        x4 = self.down3(x3_pool)  # [B, 512, H/4, W/4]
        x4_pool = self.pool3(x4)  # [B, 512, H/8, W/8]

        # Bridge
        x5 = self.bridge(x4_pool)  # [B, 512, H/8, W/8]

        # Decoder path with skip connections
        x = self.up1(x5, x4)  # [B, 512, H/4, W/4]
        x = self.up2(x, x3)  # [B, 256, H/2, W/2]
        x = self.up3(x, x2)  # [B, 128, H, W]

        # Output
        x = self.outc(x)  # [B, 3, H, W]

        # Global residual connection
        residual = self.global_res(input_img)
        output = torch.clamp(x + residual, 0, 1)

        return output

# Step 4: Training Infrastructure
class ImageEnhancementModule(pl.LightningModule):
    def __init__(self, lr=0.001):
        super().__init__()
        self.model = ImageEnhancerNet()
        self.lr = lr

        # Save hyperparameters
        self.save_hyperparameters()

        # Initialize LPIPS loss for perceptual quality - moved to setup method
        self.lpips_loss = None

    def setup(self, stage=None):
        # Initialize LPIPS here so it's on the correct device
        if self.lpips_loss is None:
            self.lpips_loss = lpips.LPIPS(net='alex').to(self.device)

    def forward(self, x):
        return self.model(x)

    def training_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        # 1. Pixel-wise loss (L1)
        l1_loss = F.l1_loss(outputs, hr_imgs)

        # 2. Perceptual loss (LPIPS)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # 3. TV loss for smoothness
        tv_loss = self.total_variation_loss(outputs)

        # Combined loss
        loss = l1_loss + 0.1 * perceptual_loss + 0.0001 * tv_loss

        # Log losses
        self.log('train_l1_loss', l1_loss)
        self.log('train_perceptual_loss', perceptual_loss)
        self.log('train_tv_loss', tv_loss)
        self.log('train_total_loss', loss)

        return loss

    def validation_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        l1_loss = F.l1_loss(outputs, hr_imgs)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # Calculate metrics
        psnr_val = self.psnr(outputs, hr_imgs)
        ssim_val = self.ssim(outputs, hr_imgs)

        # Log metrics
        self.log('val_l1_loss', l1_loss, sync_dist=True)
        self.log('val_perceptual_loss', perceptual_loss, sync_dist=True)
        self.log('val_psnr', psnr_val, sync_dist=True)
        self.log('val_ssim', ssim_val, sync_dist=True)

        # Visualize results occasionally
        if batch_idx == 0:
            # Save sample images
            self.save_images(lr_imgs, outputs, hr_imgs)

        return {'val_loss': l1_loss, 'val_psnr': psnr_val, 'val_ssim': ssim_val}

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.lr)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=5, verbose=True
        )

        return {
            'optimizer': optimizer,
            'lr_scheduler': {
                'scheduler': scheduler,
                'monitor': 'val_l1_loss',
                'interval': 'epoch',
                'frequency': 1
            }
        }

    def total_variation_loss(self, x):
        # Total variation loss for smoother outputs
        batch_size = x.size()[0]
        h_x = x.size()[2]
        w_x = x.size()[3]
        count_h = self._tensor_size(x[:, :, 1:, :])
        count_w = self._tensor_size(x[:, :, :, 1:])
        h_tv = torch.pow((x[:, :, 1:, :] - x[:, :, :h_x-1, :]), 2).sum()
        w_tv = torch.pow((x[:, :, :, 1:] - x[:, :, :, :w_x-1]), 2).sum()
        return (h_tv + w_tv) / (batch_size * count_h * count_w)

    def _tensor_size(self, t):
        return t.size()[1] * t.size()[2] * t.size()[3]

    def psnr(self, img1, img2):
        # Peak Signal-to-Noise Ratio
        mse = F.mse_loss(img1, img2)
        if mse < 1e-10:
            return torch.tensor(100.0).to(mse.device)
        return 20 * torch.log10(1.0 / torch.sqrt(mse))

    def ssim(self, img1, img2):
        # Simplified Structural Similarity Index
        C1 = 0.01**2
        C2 = 0.03**2

        img1 = img1.view(img1.size(0), -1)
        img2 = img2.view(img2.size(0), -1)

        mu1 = img1.mean(dim=1)
        mu2 = img2.mean(dim=1)

        sigma1 = (img1 - mu1.unsqueeze(1)).pow(2).mean(dim=1)
        sigma2 = (img2 - mu2.unsqueeze(1)).pow(2).mean(dim=1)
        sigma12 = ((img1 - mu1.unsqueeze(1)) * (img2 - mu2.unsqueeze(1))).mean(dim=1)

        mu1 = mu1.unsqueeze(1)
        mu2 = mu2.unsqueeze(1)

        ssim_value = ((2 * mu1 * mu2 + C1) * (2 * sigma12 + C2)) / \
                    ((mu1.pow(2) + mu2.pow(2) + C1) * (sigma1 + sigma2 + C2))

        return ssim_value.mean()

    # def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
    #     # Create a grid of images
    #     num_samples = min(lr_imgs.size(0), max_samples)

    #     # Convert tensors to numpy arrays
    #     lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
    #     sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
    #     hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

    #     # Create figure
    #     plt.figure(figsize=(15, 5 * num_samples))

    #     for i in range(num_samples):
    #         # Original low-res
    #         plt.subplot(num_samples, 3, i*3 + 1)
    #         plt.imshow(np.clip(lr_imgs_np[i], 0, 1))
    #         plt.title('Low Resolution')
    #         plt.axis('off')

    #         # Enhanced
    #         plt.subplot(num_samples, 3, i*3 + 2)
    #         plt.imshow(np.clip(sr_imgs_np[i], 0, 1))
    #         plt.title('Enhanced')
    #         plt.axis('off')

    #         # Ground truth
    #         plt.subplot(num_samples, 3, i*3 + 3)
    #         plt.imshow(np.clip(hr_imgs_np[i], 0, 1))
    #         plt.title('High Resolution')
    #         plt.axis('off')

    #     # Save figure
    #     os.makedirs('results', exist_ok=True)
    #     plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
    #     plt.close()
    def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
    # Create a grid of images
      num_samples = min(lr_imgs.size(0), max_samples)

      # Convert tensors to numpy arrays
      lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
      sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
      hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

      # Ensure the images are in the range [0, 1] and correct dtype
      lr_imgs_np = np.clip(lr_imgs_np, 0, 1).astype(np.float32)
      sr_imgs_np = np.clip(sr_imgs_np, 0, 1).astype(np.float32)
      hr_imgs_np = np.clip(hr_imgs_np, 0, 1).astype(np.float32)

      # Create figure
      plt.figure(figsize=(15, 5 * num_samples))

      for i in range(num_samples):
          # Original low-res
          plt.subplot(num_samples, 3, i*3 + 1)
          plt.imshow(lr_imgs_np[i])
          plt.title('Low Resolution')
          plt.axis('off')

          # Enhanced
          plt.subplot(num_samples, 3, i*3 + 2)
          plt.imshow(sr_imgs_np[i])
          plt.title('Enhanced')
          plt.axis('off')

          # Ground truth
          plt.subplot(num_samples, 3, i*3 + 3)
          plt.imshow(hr_imgs_np[i])
          plt.title('High Resolution')
          plt.axis('off')

      # Save figure
      os.makedirs('results', exist_ok=True)
      plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
      plt.close()


# Main execution
def main():
    print("Starting image enhancement AI training...")

    # Download and prepare dataset
    try:
        data_dir = download_div2k()
    except Exception as e:
        print(f"Error downloading dataset: {e}")
        print("Checking if data directory already exists...")
        if os.path.exists("data/DIV2K_valid_HR"):
            data_dir = "data/DIV2K_valid_HR/"
            print(f"Using existing data directory: {data_dir}")
        else:
            raise RuntimeError("Failed to download dataset and no existing data found.")

    # Create data module
    data_module = ImageEnhancementDataModule(
        data_dir=data_dir,
        batch_size=4,  # Adjust based on your GPU memory
        img_size=256
    )

    # Initialize model
    model = ImageEnhancementModule(lr=0.0002)

    # Setup callbacks
    checkpoint_callback = ModelCheckpoint(
        dirpath='models',
        filename='image_enhancer-{epoch:02d}-{val_psnr:.2f}',
        save_top_k=3,
        monitor='val_psnr',
        mode='max'
    )

    early_stop_callback = EarlyStopping(
        monitor='val_psnr',
        patience=10,
        mode='max'
    )

    logger = TensorBoardLogger('logs', name='image_enhancer')

    # Initialize trainer
    trainer = pl.Trainer(
        accelerator='auto',  # Automatically detect GPU or CPU
        devices=1,
        max_epochs=30,
        callbacks=[checkpoint_callback, early_stop_callback],
        logger=logger,
        log_every_n_steps=10,
        precision='16-mixed' if torch.cuda.is_available() else '32'  # Use mixed precision for faster training
    )

    # Train model
    print("Starting training...")
    trainer.fit(model, data_module)

    # Save final model
    torch.save(model.state_dict(), 'models/image_enhancer_final.pth')

    print(f"Training completed. Best model saved: {checkpoint_callback.best_model_path}")
    print(f"Best PSNR: {checkpoint_callback.best_model_score:.4f}")


if __name__ == "__main__":
    main()
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping
from pytorch_lightning.loggers import TensorBoardLogger
import shutil
import numpy as np
import matplotlib.pyplot as plt
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
from tqdm.auto import tqdm
import lpips
import glob
import random
from pathlib import Path
import zipfile
import gdown
import time
from datetime import datetime
import requests

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Create directories
os.makedirs("data", exist_ok=True)
os.makedirs("models", exist_ok=True)
os.makedirs("results", exist_ok=True)

# Step 1: Define Enhancement Goals
# We'll focus on:
# - Super-resolution (upscaling)
# - Noise reduction
# - Color enhancement
# - Sharpening

# Step 2: Data Collection & Preparation
class ImageEnhancementDataset(Dataset):
    def __init__(self, hr_images_dir, transform=None, degradation_transform=None,
                 img_size=256, scale_factor=4):
        self.hr_images_dir = hr_images_dir
        self.image_paths = glob.glob(os.path.join(hr_images_dir, "**/*.jpg"), recursive=True) + \
                         glob.glob(os.path.join(hr_images_dir, "**/*.png"), recursive=True)
        self.transform = transform
        self.degradation_transform = degradation_transform
        self.img_size = img_size
        self.scale_factor = scale_factor

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]

        # Read image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Create HR image
        if self.transform:
            hr_img = self.transform(image=img)['image']
        else:
            hr_img = img

        # Create LR image with degradations
        if self.degradation_transform:
            lr_img = self.degradation_transform(image=hr_img)['image']
        else:
            # Default degradation: downscale + mild noise
            h, w = hr_img.shape[:2]
            lr_h, lr_w = h // self.scale_factor, w // self.scale_factor
            lr_img = cv2.resize(hr_img, (lr_w, lr_h), interpolation=cv2.INTER_CUBIC)

            # Add noise
            noise = np.random.normal(0, 5, lr_img.shape).astype(np.uint8)
            lr_img = np.clip(lr_img + noise, 0, 255).astype(np.uint8)

            # Upscale back to original size for paired training
            lr_img = cv2.resize(lr_img, (w, h), interpolation=cv2.INTER_CUBIC)

        # Convert to tensors
        hr_tensor = transforms.ToTensor()(hr_img)
        lr_tensor = transforms.ToTensor()(lr_img)

        return {'lr': lr_tensor, 'hr': hr_tensor, 'path': img_path}

def download_div2k():
    print("Downloading DIV2K dataset...")

    url = "https://data.vision.ee.ethz.ch/cvl/DIV2K/DIV2K_valid_HR.zip"
    output = "data/DIV2K_valid_HR.zip"

    if not os.path.exists(output):
        r = requests.get(url, stream=True)
        with open(output, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

    # Create extraction directory if it doesn't exist
    os.makedirs("data/DIV2K_valid_HR", exist_ok=True)

    with zipfile.ZipFile(output, 'r') as zip_ref:
        zip_ref.extractall("data")

    print("Dataset downloaded and extracted to data/DIV2K_valid_HR/")
    return "data/DIV2K_valid_HR/"

# Define data transforms
def get_transforms(img_size=256):
    # HR image transform
    hr_transform = A.Compose([
        A.RandomCrop(height=img_size, width=img_size),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1, p=0.5),
    ])

    # LR image degradation transform (simulates various image problems)
    degradation_transform = A.Compose([
        # Downsample and upsample to lose information
        A.OneOf([
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_NEAREST, p=0.5),
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_LINEAR, p=0.5),
        ], p=1.0),

        # Add noise
        A.OneOf([
            A.GaussNoise(var_limit=(10.0, 50.0), p=0.5),
            A.ISONoise(color_shift=(0.01, 0.05), intensity=(0.1, 0.5), p=0.5),
            A.MultiplicativeNoise(multiplier=(0.9, 1.1), p=0.3),
        ], p=0.8),

        # Blur
        A.OneOf([
            A.GaussianBlur(blur_limit=(3, 7), p=0.5),
            A.MotionBlur(blur_limit=(3, 7), p=0.5),
        ], p=0.5),

        # JPEG compression artifacts
        A.ImageCompression(quality_lower=50, quality_upper=90, p=0.5),

        # Color distortions
        A.OneOf([
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
            A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
        ], p=0.5),
    ])

    return hr_transform, degradation_transform

# Data preparation class
class ImageEnhancementDataModule(pl.LightningDataModule):
    def __init__(self, data_dir, batch_size=8, img_size=256, num_workers=2):
        super().__init__()
        self.data_dir = data_dir
        self.batch_size = batch_size
        self.img_size = img_size
        self.num_workers = num_workers

    def setup(self, stage=None):
        hr_transform, degradation_transform = get_transforms(self.img_size)

        # Split into train and validation
        all_images = glob.glob(os.path.join(self.data_dir, "**/*.png"), recursive=True)
        if len(all_images) == 0:  # Try jpg if no png found
            all_images = glob.glob(os.path.join(self.data_dir, "**/*.jpg"), recursive=True)

        if len(all_images) == 0:
            raise ValueError(f"No images found in {self.data_dir}. Please check the path.")

        random.shuffle(all_images)
        split_idx = int(len(all_images) * 0.8)

        # Create train and validation directories
        os.makedirs("data/train", exist_ok=True)
        os.makedirs("data/val", exist_ok=True)

        # Create copies for train/val split
        for i, img_path in enumerate(all_images[:split_idx]):
            shutil.copy2(img_path, f"data/train/{i}.jpg")

        for i, img_path in enumerate(all_images[split_idx:]):
            shutil.copy2(img_path, f"data/val/{i}.jpg")

        self.train_dataset = ImageEnhancementDataset(
            "data/train",
            transform=hr_transform,
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

        self.val_dataset = ImageEnhancementDataset(
            "data/val",
            transform=A.Compose([A.RandomCrop(height=self.img_size, width=self.img_size)]),
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

    def train_dataloader(self):
        return DataLoader(
            self.train_dataset,
            batch_size=self.batch_size,
            shuffle=True,
            num_workers=self.num_workers,
            pin_memory=True
        )

    def val_dataloader(self):
        return DataLoader(
            self.val_dataset,
            batch_size=self.batch_size,
            shuffle=False,
            num_workers=self.num_workers,
            pin_memory=True
        )

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels=None):
        super(ResidualBlock, self).__init__()
        if out_channels is None:
            out_channels = in_channels

        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)

        # Projection to match input to output shape if needed
        self.projection = nn.Sequential()
        if in_channels != out_channels:
            self.projection = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(out_channels)
            )

    def forward(self, x):
        residual = self.projection(x)  # Project input to match output dimensions if needed

        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))

        out += residual  # Add residual connection
        out = self.relu(out)

        return out

class ChannelAttention(nn.Module):
    def __init__(self, channels, reduction=16):
        super(ChannelAttention, self).__init__()
        # Ensure reduction doesn't make channels too small
        reduction = min(reduction, max(1, channels // 4))

        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc = nn.Sequential(
            nn.Conv2d(channels, max(1, channels // reduction), 1, bias=False),
            nn.ReLU(inplace=True),
            nn.Conv2d(max(1, channels // reduction), channels, 1, bias=False)
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = self.fc(self.avg_pool(x))
        max_out = self.fc(self.max_pool(x))
        out = avg_out + max_out
        return self.sigmoid(out) * x

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size=kernel_size, padding=kernel_size//2)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        out = torch.cat([avg_out, max_out], dim=1)
        out = self.conv(out)
        return self.sigmoid(out) * x

class DownBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DownBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(in_channels=out_channels, out_channels=out_channels)
        self.channel_attention = ChannelAttention(out_channels)
        self.spatial_attention = SpatialAttention()

    def forward(self, x):
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        x = self.spatial_attention(x)
        return x

class UpBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UpBlock, self).__init__()
        self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        # Corrected: input channels should be in_channels + skip connection's channels
        self.conv = nn.Conv2d(in_channels + out_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(out_channels)
        self.channel_attention = ChannelAttention(out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # Now ensure x1 has the same spatial dimensions as x2
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                         diffY // 2, diffY - diffY // 2])

        # Concatenate along channel dimension
        x = torch.cat([x2, x1], dim=1)
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        return x

class ImageEnhancerNet(nn.Module):
    def __init__(self):
        super(ImageEnhancerNet, self).__init__()
        # Initial convolution to extract features
        self.inc = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Encoder (downsampling) path
        self.down1 = DownBlock(64, 128)
        self.pool1 = nn.MaxPool2d(2)
        self.down2 = DownBlock(128, 256)
        self.pool2 = nn.MaxPool2d(2)
        self.down3 = DownBlock(256, 512)
        self.pool3 = nn.MaxPool2d(2)

        # Bridge
        self.bridge = nn.Sequential(
            ResidualBlock(512, 512),  # Keep the channels consistent
            ResidualBlock(512, 512),
            ChannelAttention(512)
        )

        # Decoder (upsampling) path - fixed channel dimensions
        self.up1 = UpBlock(512, 512)  # Bridge output + skip connection from down3
        self.up2 = UpBlock(512, 256)  # up1 output + skip connection from down2
        self.up3 = UpBlock(256, 128)  # up2 output + skip connection from down1

        # Final layers to produce the output
        self.outc = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 3, kernel_size=3, padding=1),
            nn.Sigmoid()  # Output values between 0 and 1
        )

        # Global residual learning
        self.global_res = nn.Conv2d(3, 3, kernel_size=1)

    def forward(self, x):
        # Save input for global residual connection
        input_img = x

        # Encoder path
        x1 = self.inc(x)  # [B, 64, H, W]

        x2 = self.down1(x1)  # [B, 128, H, W]
        x2_pool = self.pool1(x2)  # [B, 128, H/2, W/2]

        x3 = self.down2(x2_pool)  # [B, 256, H/2, W/2]
        x3_pool = self.pool2(x3)  # [B, 256, H/4, W/4]

        x4 = self.down3(x3_pool)  # [B, 512, H/4, W/4]
        x4_pool = self.pool3(x4)  # [B, 512, H/8, W/8]

        # Bridge
        x5 = self.bridge(x4_pool)  # [B, 512, H/8, W/8]

        # Decoder path with skip connections
        x = self.up1(x5, x4)  # [B, 512, H/4, W/4]
        x = self.up2(x, x3)  # [B, 256, H/2, W/2]
        x = self.up3(x, x2)  # [B, 128, H, W]

        # Output
        x = self.outc(x)  # [B, 3, H, W]

        # Global residual connection
        residual = self.global_res(input_img)
        output = torch.clamp(x + residual, 0, 1)

        return output

# Step 4: Training Infrastructure
class ImageEnhancementModule(pl.LightningModule):
    def __init__(self, lr=0.001):
        super().__init__()
        self.model = ImageEnhancerNet()
        self.lr = lr

        # Save hyperparameters
        self.save_hyperparameters()

        # Initialize LPIPS loss for perceptual quality - moved to setup method
        self.lpips_loss = None

    def setup(self, stage=None):
        # Initialize LPIPS here so it's on the correct device
        if self.lpips_loss is None:
            self.lpips_loss = lpips.LPIPS(net='alex').to(self.device)

    def forward(self, x):
        return self.model(x)

    def training_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        # 1. Pixel-wise loss (L1)
        l1_loss = F.l1_loss(outputs, hr_imgs)

        # 2. Perceptual loss (LPIPS)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # 3. TV loss for smoothness
        tv_loss = self.total_variation_loss(outputs)

        # Combined loss
        loss = l1_loss + 0.1 * perceptual_loss + 0.0001 * tv_loss

        # Log losses
        self.log('train_l1_loss', l1_loss)
        self.log('train_perceptual_loss', perceptual_loss)
        self.log('train_tv_loss', tv_loss)
        self.log('train_total_loss', loss)

        return loss

    def validation_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        l1_loss = F.l1_loss(outputs, hr_imgs)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # Calculate metrics
        psnr_val = self.psnr(outputs, hr_imgs)
        ssim_val = self.ssim(outputs, hr_imgs)

        # Log metrics
        self.log('val_l1_loss', l1_loss, sync_dist=True)
        self.log('val_perceptual_loss', perceptual_loss, sync_dist=True)
        self.log('val_psnr', psnr_val, sync_dist=True)
        self.log('val_ssim', ssim_val, sync_dist=True)

        # Visualize results occasionally
        if batch_idx == 0:
            # Save sample images
            self.save_images(lr_imgs, outputs, hr_imgs)

        return {'val_loss': l1_loss, 'val_psnr': psnr_val, 'val_ssim': ssim_val}

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.lr)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=5, verbose=True
        )

        return {
            'optimizer': optimizer,
            'lr_scheduler': {
                'scheduler': scheduler,
                'monitor': 'val_l1_loss',
                'interval': 'epoch',
                'frequency': 1
            }
        }

    def total_variation_loss(self, x):
        # Total variation loss for smoother outputs
        batch_size = x.size()[0]
        h_x = x.size()[2]
        w_x = x.size()[3]
        count_h = self._tensor_size(x[:, :, 1:, :])
        count_w = self._tensor_size(x[:, :, :, 1:])
        h_tv = torch.pow((x[:, :, 1:, :] - x[:, :, :h_x-1, :]), 2).sum()
        w_tv = torch.pow((x[:, :, :, 1:] - x[:, :, :, :w_x-1]), 2).sum()
        return (h_tv + w_tv) / (batch_size * count_h * count_w)

    def _tensor_size(self, t):
        return t.size()[1] * t.size()[2] * t.size()[3]

    def psnr(self, img1, img2):
        # Peak Signal-to-Noise Ratio
        mse = F.mse_loss(img1, img2)
        if mse < 1e-10:
            return torch.tensor(100.0).to(mse.device)
        return 20 * torch.log10(1.0 / torch.sqrt(mse))

    def ssim(self, img1, img2):
        # Simplified Structural Similarity Index
        C1 = 0.01**2
        C2 = 0.03**2

        img1 = img1.view(img1.size(0), -1)
        img2 = img2.view(img2.size(0), -1)

        mu1 = img1.mean(dim=1)
        mu2 = img2.mean(dim=1)

        sigma1 = (img1 - mu1.unsqueeze(1)).pow(2).mean(dim=1)
        sigma2 = (img2 - mu2.unsqueeze(1)).pow(2).mean(dim=1)
        sigma12 = ((img1 - mu1.unsqueeze(1)) * (img2 - mu2.unsqueeze(1))).mean(dim=1)

        mu1 = mu1.unsqueeze(1)
        mu2 = mu2.unsqueeze(1)

        ssim_value = ((2 * mu1 * mu2 + C1) * (2 * sigma12 + C2)) / \
                    ((mu1.pow(2) + mu2.pow(2) + C1) * (sigma1 + sigma2 + C2))

        return ssim_value.mean()

    # def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
    #     # Create a grid of images
    #     num_samples = min(lr_imgs.size(0), max_samples)

    #     # Convert tensors to numpy arrays
    #     lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
    #     sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
    #     hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

    #     # Create figure
    #     plt.figure(figsize=(15, 5 * num_samples))

    #     for i in range(num_samples):
    #         # Original low-res
    #         plt.subplot(num_samples, 3, i*3 + 1)
    #         plt.imshow(np.clip(lr_imgs_np[i], 0, 1))
    #         plt.title('Low Resolution')
    #         plt.axis('off')

    #         # Enhanced
    #         plt.subplot(num_samples, 3, i*3 + 2)
    #         plt.imshow(np.clip(sr_imgs_np[i], 0, 1))
    #         plt.title('Enhanced')
    #         plt.axis('off')

    #         # Ground truth
    #         plt.subplot(num_samples, 3, i*3 + 3)
    #         plt.imshow(np.clip(hr_imgs_np[i], 0, 1))
    #         plt.title('High Resolution')
    #         plt.axis('off')

    #     # Save figure
    #     os.makedirs('results', exist_ok=True)
    #     plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
    #     plt.close()
    def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
    # Create a grid of images
      num_samples = min(lr_imgs.size(0), max_samples)

      # Convert tensors to numpy arrays
      lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
      sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
      hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

      # Ensure the images are in the range [0, 1] and correct dtype
      lr_imgs_np = np.clip(lr_imgs_np, 0, 1).astype(np.float32)
      sr_imgs_np = np.clip(sr_imgs_np, 0, 1).astype(np.float32)
      hr_imgs_np = np.clip(hr_imgs_np, 0, 1).astype(np.float32)

      # Create figure
      plt.figure(figsize=(15, 5 * num_samples))

      for i in range(num_samples):
          # Original low-res
          plt.subplot(num_samples, 3, i*3 + 1)
          plt.imshow(lr_imgs_np[i])
          plt.title('Low Resolution')
          plt.axis('off')

          # Enhanced
          plt.subplot(num_samples, 3, i*3 + 2)
          plt.imshow(sr_imgs_np[i])
          plt.title('Enhanced')
          plt.axis('off')

          # Ground truth
          plt.subplot(num_samples, 3, i*3 + 3)
          plt.imshow(hr_imgs_np[i])
          plt.title('High Resolution')
          plt.axis('off')

      # Save figure
      os.makedirs('results', exist_ok=True)
      plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
      plt.close()


# Main execution
def main():

    # Create data module
    data_module = ImageEnhancementDataModule(
        data_dir="/content/data/DIV2K_valid_HR",
        batch_size=4,  # Adjust based on your GPU memory
        img_size=256
    )

    # Initialize model
    model = ImageEnhancementModule(lr=0.0002)
    state_dict = torch.load('/content/models/image_enhancer_final.pth', map_location='cpu')
    model.load_state_dict(state_dict, strict=False)
    model.eval()


    test_model(model, data_module)






def test_model(model, data_module):
    """Test the trained model on a few sample images."""
    print("Testing the model on sample images...")

    # Load a batch from validation set
    val_dataloader = data_module.val_dataloader()

    # Ensure the dataloader has data
    if len(val_dataloader) == 0:
        print("No validation data available. Skipping testing.")
        return

    test_batch = next(iter(val_dataloader))

    # Move to GPU if available
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    model.eval()

    lr_imgs = test_batch['lr'].to(device)
    hr_imgs = test_batch['hr'].to(device)

    # Generate enhanced images
    with torch.no_grad():
        enhanced_imgs = model(lr_imgs)

    # Calculate metrics
    psnr_values = []
    ssim_values = []

    for i in range(lr_imgs.size(0)):
        psnr_val = model.psnr(enhanced_imgs[i:i+1], hr_imgs[i:i+1]).item()
        ssim_val = model.ssim(enhanced_imgs[i:i+1], hr_imgs[i:i+1]).item()

        psnr_values.append(psnr_val)
        ssim_values.append(ssim_val)

    avg_psnr = sum(psnr_values) / len(psnr_values)
    avg_ssim = sum(ssim_values) / len(ssim_values)

    print(f"Test Results - Average PSNR: {avg_psnr:.4f}, Average SSIM: {avg_ssim:.4f}")

    # Save sample enhanced images
    os.makedirs('results/test', exist_ok=True)

    for i in range(min(4, lr_imgs.size(0))):
        lr_img = lr_imgs[i].cpu().detach().permute(1, 2, 0).numpy()
        sr_img = enhanced_imgs[i].cpu().detach().permute(1, 2, 0).numpy()
        hr_img = hr_imgs[i].cpu().detach().permute(1, 2, 0).numpy()

        # Clip and convert to float32
        lr_img = np.clip(lr_img, 0, 1).astype(np.float32)
        sr_img = np.clip(sr_img, 0, 1).astype(np.float32)
        hr_img = np.clip(hr_img, 0, 1).astype(np.float32)

        # Plot side-by-side
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 3, 1)
        plt.imshow(lr_img)
        plt.title("Low Resolution")
        plt.axis("off")

        plt.subplot(1, 3, 2)
        plt.imshow(sr_img)
        plt.title("Enhanced")
        plt.axis("off")

        plt.subplot(1, 3, 3)
        plt.imshow(hr_img)
        plt.title("High Resolution")
        plt.axis("off")

        plt.savefig(f"results/test/sample_{i}.png")
        plt.close()



if __name__ == "__main__":
    main()
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping
from pytorch_lightning.loggers import TensorBoardLogger
import shutil
import numpy as np
import matplotlib.pyplot as plt
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
from tqdm.auto import tqdm
import lpips
import glob
import random
from pathlib import Path
import zipfile
import gdown
import time
from datetime import datetime
import requests

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Create directories
os.makedirs("data", exist_ok=True)
os.makedirs("models", exist_ok=True)
os.makedirs("results", exist_ok=True)

# Step 1: Define Enhancement Goals
# We'll focus on:
# - Super-resolution (upscaling)
# - Noise reduction
# - Color enhancement
# - Sharpening

# Step 2: Data Collection & Preparation
class ImageEnhancementDataset(Dataset):
    def __init__(self, hr_images_dir, transform=None, degradation_transform=None,
                 img_size=256, scale_factor=4):
        self.hr_images_dir = hr_images_dir
        self.image_paths = glob.glob(os.path.join(hr_images_dir, "**/*.jpg"), recursive=True) + \
                         glob.glob(os.path.join(hr_images_dir, "**/*.png"), recursive=True)
        self.transform = transform
        self.degradation_transform = degradation_transform
        self.img_size = img_size
        self.scale_factor = scale_factor

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]

        # Read image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Create HR image
        if self.transform:
            hr_img = self.transform(image=img)['image']
        else:
            hr_img = img

        # Create LR image with degradations
        if self.degradation_transform:
            lr_img = self.degradation_transform(image=hr_img)['image']
        else:
            # Default degradation: downscale + mild noise
            h, w = hr_img.shape[:2]
            lr_h, lr_w = h // self.scale_factor, w // self.scale_factor
            lr_img = cv2.resize(hr_img, (lr_w, lr_h), interpolation=cv2.INTER_CUBIC)

            # Add noise
            noise = np.random.normal(0, 5, lr_img.shape).astype(np.uint8)
            lr_img = np.clip(lr_img + noise, 0, 255).astype(np.uint8)

            # Upscale back to original size for paired training
            lr_img = cv2.resize(lr_img, (w, h), interpolation=cv2.INTER_CUBIC)

        # Convert to tensors
        hr_tensor = transforms.ToTensor()(hr_img)
        lr_tensor = transforms.ToTensor()(lr_img)

        return {'lr': lr_tensor, 'hr': hr_tensor, 'path': img_path}

def download_div2k():
    print("Downloading DIV2K dataset...")

    url = "https://data.vision.ee.ethz.ch/cvl/DIV2K/DIV2K_valid_HR.zip"
    output = "data/DIV2K_valid_HR.zip"

    if not os.path.exists(output):
        r = requests.get(url, stream=True)
        with open(output, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

    # Create extraction directory if it doesn't exist
    os.makedirs("data/DIV2K_valid_HR", exist_ok=True)

    with zipfile.ZipFile(output, 'r') as zip_ref:
        zip_ref.extractall("data")

    print("Dataset downloaded and extracted to data/DIV2K_valid_HR/")
    return "data/DIV2K_valid_HR/"

# Define data transforms
def get_transforms(img_size=256):
    # HR image transform
    hr_transform = A.Compose([
        A.RandomCrop(height=img_size, width=img_size),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1, p=0.5),
    ])

    # LR image degradation transform (simulates various image problems)
    degradation_transform = A.Compose([
        # Downsample and upsample to lose information
        A.OneOf([
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_NEAREST, p=0.5),
            A.Downscale(scale_min=0.25, scale_max=0.75, interpolation=cv2.INTER_LINEAR, p=0.5),
        ], p=1.0),

        # Add noise
        A.OneOf([
            A.GaussNoise(var_limit=(10.0, 50.0), p=0.5),
            A.ISONoise(color_shift=(0.01, 0.05), intensity=(0.1, 0.5), p=0.5),
            A.MultiplicativeNoise(multiplier=(0.9, 1.1), p=0.3),
        ], p=0.8),

        # Blur
        A.OneOf([
            A.GaussianBlur(blur_limit=(3, 7), p=0.5),
            A.MotionBlur(blur_limit=(3, 7), p=0.5),
        ], p=0.5),

        # JPEG compression artifacts
        A.ImageCompression(quality_lower=50, quality_upper=90, p=0.5),

        # Color distortions
        A.OneOf([
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
            A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
        ], p=0.5),
    ])

    return hr_transform, degradation_transform

# Data preparation class
class ImageEnhancementDataModule(pl.LightningDataModule):
    def __init__(self, data_dir, batch_size=8, img_size=256, num_workers=2):
        super().__init__()
        self.data_dir = data_dir
        self.batch_size = batch_size
        self.img_size = img_size
        self.num_workers = num_workers

    def setup(self, stage=None):
        hr_transform, degradation_transform = get_transforms(self.img_size)

        # Split into train and validation
        all_images = glob.glob(os.path.join(self.data_dir, "**/*.png"), recursive=True)
        if len(all_images) == 0:  # Try jpg if no png found
            all_images = glob.glob(os.path.join(self.data_dir, "**/*.jpg"), recursive=True)

        if len(all_images) == 0:
            raise ValueError(f"No images found in {self.data_dir}. Please check the path.")

        random.shuffle(all_images)
        split_idx = int(len(all_images) * 0.8)

        # Create train and validation directories
        os.makedirs("data/train", exist_ok=True)
        os.makedirs("data/val", exist_ok=True)

        # Create copies for train/val split
        for i, img_path in enumerate(all_images[:split_idx]):
            shutil.copy2(img_path, f"data/train/{i}.jpg")

        for i, img_path in enumerate(all_images[split_idx:]):
            shutil.copy2(img_path, f"data/val/{i}.jpg")

        self.train_dataset = ImageEnhancementDataset(
            "data/train",
            transform=hr_transform,
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

        self.val_dataset = ImageEnhancementDataset(
            "data/val",
            transform=A.Compose([A.RandomCrop(height=self.img_size, width=self.img_size)]),
            degradation_transform=degradation_transform,
            img_size=self.img_size
        )

    def train_dataloader(self):
        return DataLoader(
            self.train_dataset,
            batch_size=self.batch_size,
            shuffle=True,
            num_workers=self.num_workers,
            pin_memory=True
        )

    def val_dataloader(self):
        return DataLoader(
            self.val_dataset,
            batch_size=self.batch_size,
            shuffle=False,
            num_workers=self.num_workers,
            pin_memory=True
        )

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels=None):
        super(ResidualBlock, self).__init__()
        if out_channels is None:
            out_channels = in_channels

        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, stride=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)

        # Projection to match input to output shape if needed
        self.projection = nn.Sequential()
        if in_channels != out_channels:
            self.projection = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(out_channels)
            )

    def forward(self, x):
        residual = self.projection(x)  # Project input to match output dimensions if needed

        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))

        out += residual  # Add residual connection
        out = self.relu(out)

        return out

class ChannelAttention(nn.Module):
    def __init__(self, channels, reduction=16):
        super(ChannelAttention, self).__init__()
        # Ensure reduction doesn't make channels too small
        reduction = min(reduction, max(1, channels // 4))

        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc = nn.Sequential(
            nn.Conv2d(channels, max(1, channels // reduction), 1, bias=False),
            nn.ReLU(inplace=True),
            nn.Conv2d(max(1, channels // reduction), channels, 1, bias=False)
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = self.fc(self.avg_pool(x))
        max_out = self.fc(self.max_pool(x))
        out = avg_out + max_out
        return self.sigmoid(out) * x

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size=kernel_size, padding=kernel_size//2)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        out = torch.cat([avg_out, max_out], dim=1)
        out = self.conv(out)
        return self.sigmoid(out) * x

class DownBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DownBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(in_channels=out_channels, out_channels=out_channels)
        self.channel_attention = ChannelAttention(out_channels)
        self.spatial_attention = SpatialAttention()

    def forward(self, x):
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        x = self.spatial_attention(x)
        return x

class UpBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UpBlock, self).__init__()
        self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        # Corrected: input channels should be in_channels + skip connection's channels
        self.conv = nn.Conv2d(in_channels + out_channels, out_channels, kernel_size=3, padding=1)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.res_block = ResidualBlock(out_channels)
        self.channel_attention = ChannelAttention(out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # Now ensure x1 has the same spatial dimensions as x2
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                         diffY // 2, diffY - diffY // 2])

        # Concatenate along channel dimension
        x = torch.cat([x2, x1], dim=1)
        x = self.relu(self.bn(self.conv(x)))
        x = self.res_block(x)
        x = self.channel_attention(x)
        return x

class ImageEnhancerNet(nn.Module):
    def __init__(self):
        super(ImageEnhancerNet, self).__init__()
        # Initial convolution to extract features
        self.inc = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Encoder (downsampling) path
        self.down1 = DownBlock(64, 128)
        self.pool1 = nn.MaxPool2d(2)
        self.down2 = DownBlock(128, 256)
        self.pool2 = nn.MaxPool2d(2)
        self.down3 = DownBlock(256, 512)
        self.pool3 = nn.MaxPool2d(2)

        # Bridge
        self.bridge = nn.Sequential(
            ResidualBlock(512, 512),  # Keep the channels consistent
            ResidualBlock(512, 512),
            ChannelAttention(512)
        )

        # Decoder (upsampling) path - fixed channel dimensions
        self.up1 = UpBlock(512, 512)  # Bridge output + skip connection from down3
        self.up2 = UpBlock(512, 256)  # up1 output + skip connection from down2
        self.up3 = UpBlock(256, 128)  # up2 output + skip connection from down1

        # Final layers to produce the output
        self.outc = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 3, kernel_size=3, padding=1),
            nn.Sigmoid()  # Output values between 0 and 1
        )

        # Global residual learning
        self.global_res = nn.Conv2d(3, 3, kernel_size=1)

    def forward(self, x):
        # Save input for global residual connection
        input_img = x

        # Encoder path
        x1 = self.inc(x)  # [B, 64, H, W]

        x2 = self.down1(x1)  # [B, 128, H, W]
        x2_pool = self.pool1(x2)  # [B, 128, H/2, W/2]

        x3 = self.down2(x2_pool)  # [B, 256, H/2, W/2]
        x3_pool = self.pool2(x3)  # [B, 256, H/4, W/4]

        x4 = self.down3(x3_pool)  # [B, 512, H/4, W/4]
        x4_pool = self.pool3(x4)  # [B, 512, H/8, W/8]

        # Bridge
        x5 = self.bridge(x4_pool)  # [B, 512, H/8, W/8]

        # Decoder path with skip connections
        x = self.up1(x5, x4)  # [B, 512, H/4, W/4]
        x = self.up2(x, x3)  # [B, 256, H/2, W/2]
        x = self.up3(x, x2)  # [B, 128, H, W]

        # Output
        x = self.outc(x)  # [B, 3, H, W]

        # Global residual connection
        residual = self.global_res(input_img)
        output = torch.clamp(x + residual, 0, 1)

        return output

# Step 4: Training Infrastructure
class ImageEnhancementModule(pl.LightningModule):
    def __init__(self, lr=0.001):
        super().__init__()
        self.model = ImageEnhancerNet()
        self.lr = lr

        # Save hyperparameters
        self.save_hyperparameters()

        # Initialize LPIPS loss for perceptual quality - moved to setup method
        self.lpips_loss = None

    def setup(self, stage=None):
        # Initialize LPIPS here so it's on the correct device
        if self.lpips_loss is None:
            self.lpips_loss = lpips.LPIPS(net='alex').to(self.device)

    def forward(self, x):
        return self.model(x)

    def training_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        # 1. Pixel-wise loss (L1)
        l1_loss = F.l1_loss(outputs, hr_imgs)

        # 2. Perceptual loss (LPIPS)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # 3. TV loss for smoothness
        tv_loss = self.total_variation_loss(outputs)

        # Combined loss
        loss = l1_loss + 0.1 * perceptual_loss + 0.0001 * tv_loss

        # Log losses
        self.log('train_l1_loss', l1_loss)
        self.log('train_perceptual_loss', perceptual_loss)
        self.log('train_tv_loss', tv_loss)
        self.log('train_total_loss', loss)

        return loss

    def validation_step(self, batch, batch_idx):
        lr_imgs = batch['lr']
        hr_imgs = batch['hr']

        # Forward pass
        outputs = self(lr_imgs)

        # Calculate losses
        l1_loss = F.l1_loss(outputs, hr_imgs)
        perceptual_loss = self.lpips_loss(outputs, hr_imgs).mean()

        # Calculate metrics
        psnr_val = self.psnr(outputs, hr_imgs)
        ssim_val = self.ssim(outputs, hr_imgs)

        # Log metrics
        self.log('val_l1_loss', l1_loss, sync_dist=True)
        self.log('val_perceptual_loss', perceptual_loss, sync_dist=True)
        self.log('val_psnr', psnr_val, sync_dist=True)
        self.log('val_ssim', ssim_val, sync_dist=True)

        # Visualize results occasionally
        if batch_idx == 0:
            # Save sample images
            self.save_images(lr_imgs, outputs, hr_imgs)

        return {'val_loss': l1_loss, 'val_psnr': psnr_val, 'val_ssim': ssim_val}

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.lr)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=5, verbose=True
        )

        return {
            'optimizer': optimizer,
            'lr_scheduler': {
                'scheduler': scheduler,
                'monitor': 'val_l1_loss',
                'interval': 'epoch',
                'frequency': 1
            }
        }

    def total_variation_loss(self, x):
        # Total variation loss for smoother outputs
        batch_size = x.size()[0]
        h_x = x.size()[2]
        w_x = x.size()[3]
        count_h = self._tensor_size(x[:, :, 1:, :])
        count_w = self._tensor_size(x[:, :, :, 1:])
        h_tv = torch.pow((x[:, :, 1:, :] - x[:, :, :h_x-1, :]), 2).sum()
        w_tv = torch.pow((x[:, :, :, 1:] - x[:, :, :, :w_x-1]), 2).sum()
        return (h_tv + w_tv) / (batch_size * count_h * count_w)

    def _tensor_size(self, t):
        return t.size()[1] * t.size()[2] * t.size()[3]

    def psnr(self, img1, img2):
        # Peak Signal-to-Noise Ratio
        mse = F.mse_loss(img1, img2)
        if mse < 1e-10:
            return torch.tensor(100.0).to(mse.device)
        return 20 * torch.log10(1.0 / torch.sqrt(mse))

    def ssim(self, img1, img2):
        # Simplified Structural Similarity Index
        C1 = 0.01**2
        C2 = 0.03**2

        img1 = img1.view(img1.size(0), -1)
        img2 = img2.view(img2.size(0), -1)

        mu1 = img1.mean(dim=1)
        mu2 = img2.mean(dim=1)

        sigma1 = (img1 - mu1.unsqueeze(1)).pow(2).mean(dim=1)
        sigma2 = (img2 - mu2.unsqueeze(1)).pow(2).mean(dim=1)
        sigma12 = ((img1 - mu1.unsqueeze(1)) * (img2 - mu2.unsqueeze(1))).mean(dim=1)

        mu1 = mu1.unsqueeze(1)
        mu2 = mu2.unsqueeze(1)

        ssim_value = ((2 * mu1 * mu2 + C1) * (2 * sigma12 + C2)) / \
                    ((mu1.pow(2) + mu2.pow(2) + C1) * (sigma1 + sigma2 + C2))

        return ssim_value.mean()

    # def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
    #     # Create a grid of images
    #     num_samples = min(lr_imgs.size(0), max_samples)

    #     # Convert tensors to numpy arrays
    #     lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
    #     sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
    #     hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

    #     # Create figure
    #     plt.figure(figsize=(15, 5 * num_samples))

    #     for i in range(num_samples):
    #         # Original low-res
    #         plt.subplot(num_samples, 3, i*3 + 1)
    #         plt.imshow(np.clip(lr_imgs_np[i], 0, 1))
    #         plt.title('Low Resolution')
    #         plt.axis('off')

    #         # Enhanced
    #         plt.subplot(num_samples, 3, i*3 + 2)
    #         plt.imshow(np.clip(sr_imgs_np[i], 0, 1))
    #         plt.title('Enhanced')
    #         plt.axis('off')

    #         # Ground truth
    #         plt.subplot(num_samples, 3, i*3 + 3)
    #         plt.imshow(np.clip(hr_imgs_np[i], 0, 1))
    #         plt.title('High Resolution')
    #         plt.axis('off')

    #     # Save figure
    #     os.makedirs('results', exist_ok=True)
    #     plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
    #     plt.close()
    def save_images(self, lr_imgs, sr_imgs, hr_imgs, max_samples=4):
    # Create a grid of images
      num_samples = min(lr_imgs.size(0), max_samples)

      # Convert tensors to numpy arrays
      lr_imgs_np = lr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
      sr_imgs_np = sr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)
      hr_imgs_np = hr_imgs[:num_samples].cpu().detach().numpy().transpose(0, 2, 3, 1)

      # Ensure the images are in the range [0, 1] and correct dtype
      lr_imgs_np = np.clip(lr_imgs_np, 0, 1).astype(np.float32)
      sr_imgs_np = np.clip(sr_imgs_np, 0, 1).astype(np.float32)
      hr_imgs_np = np.clip(hr_imgs_np, 0, 1).astype(np.float32)

      # Create figure
      plt.figure(figsize=(15, 5 * num_samples))

      for i in range(num_samples):
          # Original low-res
          plt.subplot(num_samples, 3, i*3 + 1)
          plt.imshow(lr_imgs_np[i])
          plt.title('Low Resolution')
          plt.axis('off')

          # Enhanced
          plt.subplot(num_samples, 3, i*3 + 2)
          plt.imshow(sr_imgs_np[i])
          plt.title('Enhanced')
          plt.axis('off')

          # Ground truth
          plt.subplot(num_samples, 3, i*3 + 3)
          plt.imshow(hr_imgs_np[i])
          plt.title('High Resolution')
          plt.axis('off')

      # Save figure
      os.makedirs('results', exist_ok=True)
      plt.savefig(f'results/samples_epoch_{self.current_epoch}.png')
      plt.close()


# Main execution
def main():
    print("Starting image enhancement AI training...")

    # Download and prepare dataset
    try:
        data_dir = download_div2k()
    except Exception as e:
        print(f"Error downloading dataset: {e}")
        print("Checking if data directory already exists...")
        if os.path.exists("data/DIV2K_valid_HR"):
            data_dir = "data/DIV2K_valid_HR/"
            print(f"Using existing data directory: {data_dir}")
        else:
            raise RuntimeError("Failed to download dataset and no existing data found.")

    # Create data module
    data_module = ImageEnhancementDataModule(
        data_dir=data_dir,
        batch_size=4,  # Adjust based on your GPU memory
        img_size=256
    )
    data_module.setup()
    # Initialize model
    model = ImageEnhancementModule()

    # Setup callbacks
    checkpoint_callback = ModelCheckpoint(
        dirpath='models',
        filename='image_enhancer-{epoch:02d}-{val_psnr:.2f}',
        save_top_k=3,
        monitor='val_psnr',
        mode='max'
    )

    early_stop_callback = EarlyStopping(
        monitor='val_psnr',
        patience=10,
        mode='max'
    )

    logger = TensorBoardLogger('logs', name='image_enhancer')

    # Initialize trainer
    trainer = pl.Trainer(
        accelerator='auto',  # Automatically detect GPU or CPU
        devices=1,
        max_epochs=30,
        callbacks=[checkpoint_callback, early_stop_callback],
        logger=logger,
        log_every_n_steps=10,
        precision='16-mixed' if torch.cuda.is_available() else '32'  # Use mixed precision for faster training
    )

    # # Train model
    # print("Starting training...")
    # trainer.fit(model, data_module)

    # # Save final model
    # torch.save(model.state_dict(), 'models/image_enhancer_final.pth')

    # print(f"Training completed. Best model saved: {checkpoint_callback.best_model_path}")
    # print(f"Best PSNR: {checkpoint_callback.best_model_score:.4f}")

    state_dict = torch.load('/content/image_enhancer_final.pth', map_location='cpu')
    model.load_state_dict(state_dict, strict=False)
    model.eval()

    test_model(model, data_module)


import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt
import torchvision.utils as vutils

def test_model(model, data_module):
    print("Testing the model on sample images...")

    val_dataloader = data_module.val_dataloader()

    if len(val_dataloader) == 0:
        print("No validation data available. Skipping testing.")
        return

    test_batch = next(iter(val_dataloader))
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    model.eval()

    lr_imgs = test_batch['lr'].to(device)
    hr_imgs = test_batch['hr'].to(device)

    with torch.no_grad():
        enhanced_imgs = model(lr_imgs)

    psnr_values = []
    ssim_values = []

    for i in range(lr_imgs.size(0)):
        psnr_val = model.psnr(enhanced_imgs[i:i+1], hr_imgs[i:i+1]).item()
        ssim_val = model.ssim(enhanced_imgs[i:i+1], hr_imgs[i:i+1]).item()
        psnr_values.append(psnr_val)
        ssim_values.append(ssim_val)

    avg_psnr = sum(psnr_values) / len(psnr_values)
    avg_ssim = sum(ssim_values) / len(ssim_values)
    print(f"Test Results - Average PSNR: {avg_psnr:.4f}, Average SSIM: {avg_ssim:.4f}")

    # Save and show sample images
    os.makedirs('results/test', exist_ok=True)

    for i in range(min(4, lr_imgs.size(0))):
        lr_img = lr_imgs[i].cpu()
        hr_img = hr_imgs[i].cpu()
        enhanced_img = enhanced_imgs[i].cpu()

        # Denormalize if needed
        def to_image(tensor):
            return TF.to_pil_image(torch.clamp(tensor, 0, 1))

        lr_pil = to_image(lr_img)
        hr_pil = to_image(hr_img)
        enhanced_pil = to_image(enhanced_img)

        # Save
        lr_pil.save(f'results/test/lr_{i}.png')
        hr_pil.save(f'results/test/hr_{i}.png')
        enhanced_pil.save(f'results/test/enhanc1ed_{i}.png')

        # Show
        fig, axes = plt.subplots(1, 3, figsize=(12, 4))
        axes[0].imshow(lr_pil)
        axes[0].set_title('Low-Res Input')
        axes[1].imshow(enhanced_pil)
        axes[1].set_title('Enhanced')
        axes[2].imshow(hr_pil)
        axes[2].set_title('High-Res Ground Truth')

        for ax in axes:
            ax.axis('off')
        plt.tight_layout()
        plt.show()





if __name__ == "__main__":
    main()
```

```python
import torch
from torchvision import transforms
import torchvision.transforms.functional as TF
from PIL import Image
import matplotlib.pyplot as plt
import torch.nn.functional as F

def testForOne(img_path):

  # Load original high-res image (Ground Truth)

  original_image = Image.open(img_path).convert('RGB')
  original_size = original_image.size  # (width, height)

  # Manually downscale the image to simulate low-res input
  low_res_image = original_image.resize((256, 256), resample=Image.BICUBIC)

  # Normalize and convert to tensor
  transform = transforms.Compose([
      transforms.ToTensor(),
      transforms.Normalize(mean=[0.5]*3, std=[0.5]*3)
  ])
  input_tensor = transform(low_res_image).unsqueeze(0)  # [1, 3, 256, 256]

  # Load model
  model = ImageEnhancementModule(lr=0.0002)
  state_dict = torch.load('/content/models/image_enhancer_final.pth', map_location='cpu')
  model.load_state_dict(state_dict, strict=False)
  model.eval()

  # Enhance
  with torch.no_grad():
      enhanced_tensor = model(input_tensor)  # [1, 3, 256, 256]

  # Upsample the enhanced output to original resolution
  enhanced_tensor = F.interpolate(enhanced_tensor, size=(original_size[1], original_size[0]), mode='bilinear', align_corners=False)

  # Convert all images to displayable format
  def to_image(tensor):
      return TF.to_pil_image(torch.clamp(tensor.squeeze(0), 0, 1))

  lr_pil = low_res_image
  hr_pil = original_image
  enhanced_pil = to_image(enhanced_tensor)

  # Show all 3 images
  fig, axes = plt.subplots(1, 3, figsize=(12, 4))
  axes[0].imshow(lr_pil)
  axes[0].set_title('Low-Res Input (Downscaled)')
  axes[1].imshow(enhanced_pil)
  axes[1].set_title('Enhanced Output')
  axes[2].imshow(hr_pil)
  axes[2].set_title('Original High-Res')

  for ax in axes:
      ax.axis('off')
  plt.tight_layout()
  plt.show()



for i in range(1,5):
  img_path = '/content/enhancer_test'+str(i)+'.jpg'
  testForOne(img_path)
```
